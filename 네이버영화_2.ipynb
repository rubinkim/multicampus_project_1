```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd
from urllib.parse import quote

import unicodedata
import re
from tqdm.notebook import tqdm
```

```python
titles, stories, males, females, age10s, age20s, age30s, age40s, age50s = [],[],[],[],[],[],[],[],[]
date='20210927'

for page in tqdm(range(1,2)):
    for i in range(3):
        try:
            chromedriver = 'chromedriver.exe'
            driver = webdriver.Chrome(chromedriver)
            driver.maximize_window()
            search_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&'
            url = f'{search_url}&date={date}&page={page}'
            driver.get(url)
            time.sleep(2)

            tits = driver.find_elements_by_css_selector('.tit5')
            tits[i].find_element_by_tag_name('a').click()
            time.sleep(2)

            driver.switch_to.window
            time.sleep(2)

            lis = driver.find_element_by_css_selector('.sub_tab_area #movieEndTabMenu').find_elements_by_tag_name('li')
            lis[0].find_element_by_tag_name('a').click()

            html_0 = driver.page_source
            soup = BeautifulSoup(html_0, 'html.parser')
            
            title = soup.select_one('.my_info .h_movie').text
            story = soup.select_one('.story_area .con_tx').text

            driver.switch_to.window
            time.sleep(2)

            lis = driver.find_element_by_css_selector('.sub_tab_area #movieEndTabMenu').find_elements_by_tag_name('li')
            lis[4].find_element_by_tag_name('a').click()
            html_4 = driver.page_source
            soup = BeautifulSoup(html_4, 'html.parser')
            male = soup.select_one('.grp_male .graph_point').text
            female = soup.select_one('.grp_female .graph_point').text
            ages = soup.select('.grp_age .graph_point')
            age10 = ages[0].text
            age20 = ages[1].text
            age30 = ages[2].text
            age40 = ages[3].text
            age50 = ages[4].text

            stories.append(story)
            males.append(male)
            females.append(female)
            age10s.append(age10)
            age20s.append(age20)
            age30s.append(age30)
            age40s.append(age40)
            age50s.append(age50)

        except:
            print(i)

titles, stories, males, females, age10s, age20s, age30s, age40s, age50s
```
