## 네이버영화 역대 평점 랭킹 2000개 

```python
titles, credits = [],[]
date='20211002'

for page in tqdm(range(1,3)):
    for i in range(5):

        chromedriver = 'chromedriver.exe'
        driver = webdriver.Chrome(chromedriver)
        driver.maximize_window()
        search_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&'
        url = f'{search_url}&date={date}&page={page}'
        driver.get(url)
        time.sleep(2)

        try:
            tits = driver.find_elements_by_css_selector('.tit5')
            title = tits[i].find_element_by_tag_name('a').text
            tits[i].find_element_by_tag_name('a').click()
            time.sleep(2)

            driver.switch_to.window
            time.sleep(2)

            driver.find_element_by_css_selector('.sub_tab_area #movieEndTabMenu .tab05.off').click()
            time.sleep(2)

            driver.switch_to.frame(driver.find_element_by_xpath("//*[@id='pointAfterListIframe']"))
            ps = driver.find_elements_by_css_selector('.score_reple p')

            txt = ''
            for p in ps:
                txt += p.text
            
            time.sleep(2)
                      
            titles.append(title)      
            credits.append(txt)      

        except:
            print(f'{page} {1} : {title}')

titles, credits
```
